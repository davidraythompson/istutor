This exercise builds basic skills in plotting and interpreting surface reflectance spectra, and in browsing large image datacubes.  We begin by importing the required software dependencies.

```python
import scipy as s
import pylab as plt
import figure_style # a local script
```

Our objective will be to locate specific material reflectances in the enclosed datacube, an overflight of the California Institute of Technology (CalTech) by NASA's Next Generation Airborne Visible InfraRed Imaging Spectrometer (AVIRIS-NG).  The conditions of the overflight were not optimal; it took place on November 7, 2017 very early in the morning.  This, combined with the time of year, meant a low sun-angle and an atypically-low signal-to-noise ratio (SNR). However, the data provide a diverse range of materials to explore.  First, we import the target signatures (i.e., radiances), represented as space-delimited columns in ASCII text format.  Each channel is associated with a wavelength, stored in `wavelengths.txt`.

```python
# Load file containing center of each spectral window
wavelengths = s.loadtxt('data/wavelengths.txt')

# Load file containing target spectra, or radiances at each wavelength
targets = s.loadtxt('data/targets.txt')
```

```python
# Explore the wavelengths array
wavelengths
```

```python
# Explore the targets/radiances array
targets
```

Let's plot one of the spectra:

```python
plt.figure()
plt.cla()
plt.plot(wavelengths, targets[:,2]) # Python arrays are 0-based, so plot the 3rd row/spectrum
plt.xlim([400,2500])
plt.ylim([0,1.0])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Surface reflectance')
plt.title('Spectrum 3');
```

<b>Question 1:</b>  What causes the large departures from the general curve near 1400 nm, 1850 nm, and 2500 nm?
    
Plot each spectrum and save the figures to the enclosed directory for reference:

```python
# Plot each spectrum and save the output to file
for i,x in enumerate(targets.T):
    plt.figure()
    plt.cla()
    plt.plot(wavelengths, x)
    plt.xlim([400,2500])
    plt.ylim([0,1.0])
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Surface reflectance')
    plt.title('Spectrum %i'%(i+1))
    plt.savefig('images/spectrum_%i.png'%i)
```
