This exercise builds basic skills in plotting and interpreting surface reflectance spectra, and in browsing large image cubes.  We begin by importing all necessary dependencies. 

```python
import scipy as s
import pylab as plt
import figure_style
```

Our objective will be to hunt down specific material reflectances in the enclosed data cube, an overflight of the California Institute of Technology by NASA's Next Generation Airborne Visible InfraRed Imaging Spectrometer (AVIRIS-NG).  The conditions of this overflight were not optimal; it took place on November 7, 2017 very early in the morning.  This, combined with the time of year, mean a low sun angle and atypically-low Signal to Noise Ratios (SNR). However, it provides a diverse range of materials to explore.  First we import the target signatures, represented as space-delimited columns in ASCII text files.  Every channel is associated with a wavelength, stored in 'wavelengths.txt'

```python
# Load target spectra 
targets = s.loadtxt('data/targets.txt')
wavelengths = s.loadtxt('data/wavelengths.txt')
```

Take a look at one of the spectra.  

```python
plt.figure()
plt.cla()
plt.plot(wavelengths, targets[:,2])
plt.xlim([400,2500])
plt.ylim([0,1.0])
plt.xlabel('Wavelength (nm)')
plt.ylabel('Surface reflectance')
plt.title('Spectrum 3');
```


<b>Question 1:</b>  What causes the large departures from the general curve near 1400 nm, 1850 nm, and 2500 nm?
    
Now, plot all spectra and save them as figure to the enclosed directory for reference. 

```python

# Plot each spectrum and save the output to a file
for i,x in enumerate(targets.T):
    
    plt.figure()
    plt.cla()
    plt.plot(wavelengths, x)
    plt.xlim([400,2500])
    plt.ylim([0,1.0])
    plt.xlabel('Wavelength (nm)')
    plt.ylabel('Surface reflectance')
    plt.title('Spectrum %i'%(i+1))
    plt.savefig('images/spectrum_%i.png'%i)
```
